#### How is the sigmoid function defined? What does it look like when plotted as a graph? What about ReLu?
Sigmoid: $$\sigma(x) := y = \frac{1}{1+e^{-x}}$$

```python
# Import matplotlib, numpy and math
import matplotlib.pyplot as plt
import numpy as np
import math
  
x = np.linspace(-10, 10, 100)
z_sigma = 1/(1 + np.exp(-x))
  
plt.plot(x, z_sigma)
plt.xlabel("x")
plt.ylabel("Sigmoid(X)")
  
plt.show()
```

ReLU: 
$$
ReLu(x) := y = \begin{cases} 
    x \text{ for } x>0 \\ 
    0 \text{ for } x \leq 0
\end{cases}
$$

```python
z_relu = np.maximum(x, 0)
plt.plot(x, z_relu)
plt.xlabel("x")
plt.ylabel("ReLU(X)")
  
plt.show()
```

#### What is the derivative of the sigmoid function? What is the derivative of ReLu?
Sigmoid Derivative:
$$\frac{dy}{dx} = \sigma(x)(1-\sigma(x))$$

```python
sigma_helper = [1-x for x in z_sigma]
z_sigma_derived = np.multiply(z_sigma, sigma_helper)
plt.plot(x, z_sigma_derived)
plt.xlabel("x")
plt.ylabel("Sigmoid'(X)")
  
plt.show()
```

Relu Derivative:
$$
\frac{dy}{dx} = \begin{cases} 
    1 \text{ for } x>0 \\ 
    0 \text{ for } x \leq 0
\end{cases}
$$

```python
z_relu_derived = [1 if x>0 else 0 for x in x]
plt.plot(x, z_relu_derived)
plt.xlabel("x")
plt.ylabel("ReLU'(X)")
  
plt.show()
```

#### Why would one choose sigmoid or ReLU over the step function when deciding on which activation function to use?
Sigmoid and ReLU have meaningfull derivatives, whereas the stepfunction does not. We are using the derivative to perform gradient descent.
