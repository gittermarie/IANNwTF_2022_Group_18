### 2 Prepare MNIST math dataset

```python
import tensorflow_datasets as tfds
import tensorflow as tf
import numpy as np
import matplotlib.pyplot as plt

(train_ds, test_ds), ds_info = tfds.load('mnist', split=['train', 'test'], as_supervised=True, with_info=True)
```

```python
def prepare_data(mnist, batch_size, condition):
    mnist = mnist.map(lambda img, target:(tf.reshape(img, (-1,)), target))
    # change datatype from unit8 to tf.float32:
    mnist = mnist.map(lambda img, target:(tf.cast(img, tf.float32), target))
    # normalize values:
    mnist = mnist.map(lambda img, target:((img/128.)-1., target))
    # zip two images together and create added up target
    zipped_mnist = tf.data.Dataset.zip((mnist.shuffle(1000), mnist.shuffle(1000)))
    if condition:
        zipped_mnist = zipped_mnist.map(lambda x, y: (x[0], y[0], x[1]-y[1]))
        zipped_mnist = zipped_mnist.map(lambda img1, img2, target:(img1, img2, tf.one_hot(target, depth = 19)))
    else:
        zipped_mnist = zipped_mnist.map(lambda x, y: (x[0], y[0], (x[1]+y[1])<5))
        zipped_mnist = zipped_mnist.map(lambda img1, img2, target : (img1, img2, [tf.cast(target, tf.float32)]))
    
    zipped_mnist = zipped_mnist.cache()                           
    zipped_mnist = zipped_mnist.batch(batch_size)
    # zipped_mnist = zipped_mnist.prefetch(10)
    
    return zipped_mnist
```

### 3 Building shared weight models

```python
from tensorflow.keras.layers import Dense

class MyModel(tf.keras.Model):
    
    def __init__(self, unitslist, condition):
        super(MyModel, self).__init__()
        # no need to give inpt dimensions for each layers weight matrix
        # keras.module will figure it out on its own, as soon as we 
        # feed it its first input.
        self.my_layers =[]
        for units in unitslist:
            self.my_layers.append(tf.keras.layers.Dense(units, activation=tf.nn.relu))
        # softmax distributes values across all 10 output options (digits 0-9) 
        # that add up to 1, kind of like a probability
        if condition:
            self.out = tf.keras.layers.Dense(19, activation=tf.nn.softmax)
        else:
            self.out = tf.keras.layers.Dense(1, activation=tf.nn.sigmoid)
        
    @tf.function
    def call(self, images1, images2):
        for layer in self.my_layers:
            images1 = layer(images1)
            images2 = layer(images2)
            
        x = self.out(tf.concat((images1, images2), axis=1))
        return x
```

### 4 Training the Network

```python
def training_loop(model, training_data, testing_data, loss_function, optimizer, condition):
    for images1, images2, targets in training_data:
        train_loss = []
        train_accuracy = []
        test_loss = []
        test_accuracy = []
        with tf.GradientTape() as tape:
            predictions = model(images1, images2)
            loss = loss_function(targets, predictions)
            if condition:
                accuracy = np.mean(np.argmax(targets, axis=1) == np.argmax(predictions, axis=1))
            else:
                accuracy = np.mean(np.absolute(predictions.numpy()-targets)<0.5)
        gradients = tape.gradient(loss, model.trainable_variables)
        optimizer.apply_gradients(zip(gradients, model.trainable_variables))
        train_loss.append(loss)
        train_accuracy.append(accuracy)
        
    for images1, images2, targets in testing_data:
        predictions = model(images1, images2)
        loss = loss_function(targets, predictions)
        if condition:
                accuracy = np.mean(np.argmax(targets, axis=1) == np.argmax(predictions, axis=1))
        else:
                accuracy = np.mean(np.absolute(predictions.numpy()-targets)<0.5)
                
        test_loss.append(loss)
        test_accuracy.append(accuracy)

    return train_loss, train_accuracy, test_loss, test_accuracy
```

```python
def train(condition, optimizer, n_epochs, batch_size):
    model = MyModel([256,256], condition)
    loss_function =  tf.keras.losses.CategoricalCrossentropy() if condition else tf.keras.losses.BinaryCrossentropy()
    training_data = prepare_data(train_ds, batch_size, condition)
    testing_data = prepare_data(test_ds, batch_size, condition)
    
    train_losses = []
    train_accuracies = []
    test_losses = []
    test_accuracies = []

    for n in range(n_epochs):
        train_loss, train_accuracy, test_loss, test_accuracy = training_loop(model, training_data, testing_data, loss_function, optimizer, condition)
        train_losses.append(np.mean(train_loss))
        train_accuracies.append(np.mean(train_accuracy))
        test_losses.append(np.mean(test_loss))
        # print(test_accuracy)
        test_accuracies.append(np.mean(test_accuracy))
        # print(f'Epoch: {n} current testing accuracy {test_accuracies[n]}')
    
    return train_losses, train_accuracies, test_losses, test_accuracies
```

### 5 Experiments

```python
def visualize(ax, title, train_losses, train_accuracies, test_losses, test_accuracies):
    line1, = ax.plot(train_losses, "b-")
    line2, = ax.plot(test_losses, "r-")
    line3, = ax.plot(train_accuracies, "b:")
    line4, = ax.plot(test_accuracies, "r:")
    ax.legend((line1, line2, line3, line4), ("training loss", "testloss", "train accuracy", "test accuracy"))
    ax.set_title(title)
```

```python
optimizers = []
optimizers.append(["SGD", tf.keras.optimizers.SGD()])
optimizers.append(["SGD with momentum", tf.keras.optimizers.SGD(momentum=0.9)])
optimizers.append(["Adam", tf.keras.optimizers.Adam()])
optimizers.append(["RMSProp", tf.keras.optimizers.RMSprop()])
optimizers.append(["AdaGrad", tf.keras.optimizers.Adagrad()])
```

#### 5.1 Subtask 01

```python
fig, axs = plt.subplots(2, 3, figsize=(15, 8), constrained_layout=True)

fig.suptitle('Training Subtask 01')

k = -1
for i, optimizer in enumerate(optimizers):
    k += 1 if i%2 == 0 else 0
    train_losses, train_accuracies, test_losses, test_accuracies = train(False, optimizer[1], 10, 32)
    visualize(axs[i%2, k], optimizer[0], train_losses, train_accuracies, test_losses, test_accuracies)
```

```python
fig, axs = plt.subplots(2, 3, figsize=(15, 8), constrained_layout=True)

fig.suptitle('Training Subtask 02')

k = -1
for i, optimizer in enumerate(optimizers):
    k += 1 if i%2 == 0 else 0
    train_losses, train_accuracies, test_losses, test_accuracies = train(True, optimizer[1], 10, 32)
    visualize(axs[i%2, k], optimizer[0], train_losses, train_accuracies, test_losses, test_accuracies)
```
